```python
import pandas as pd
```

```python
df = pd.read_csv('../data/Setka.csv')
df.head()
```

```python
def create_player_base_ratings(df):
    players = pd.concat([df['Player1'], df['Player2']]).unique()
    rows = [{'Player': player, 'Base Rating': 1500, 'Matches Played': 0, 'Wins': 0, 'Losses': 0} for player in players]
    return pd.DataFrame(rows)
```

```python
player_base_ratings = create_player_base_ratings(df)
player_base_ratings.to_csv('../data/player_base.csv', index=False)
```
